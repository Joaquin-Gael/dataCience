```python
casas_df = pd.read_csv('./casas_argentina_logico.csv')
casas_df
```

```python
casas_df['precio'] = casas_df['precio'] // 100 if casas_df['precio'].max() > 5000 else casas_df['precio']
casas_df
```

```python
casas_df.describe().round()
```

# Regrecion Lineal Simple

```python
casas_df.sort_values(by='precio', ascending=False)
```

```python
piezas_X = casas_df['piezas']
precio_Y = casas_df['precio']
```

```python
casas_df.plot(kind='scatter', x='piezas', y='precio')
```

```python
piezas_X = sm.add_constant(piezas_X)
```

```python
lm = sm.OLS(precio_Y, piezas_X).fit()
```

```python
lm.predict(piezas_X)
```

```python
lm.summary()
```

```python
# coef piezas: 341.3391
# coef const: -38.2938

# ecu lineal: y = ax + b

y_pred = (341.3391*piezas_X['piezas']) + -38.2938
y_pred
```

```python
plt.figure(figsize=(6,4),tight_layout=True)
sns.scatterplot(x=piezas_X['piezas'], y=precio_Y)

sns.lineplot(x=piezas_X['piezas'], y=y_pred, color='red')

plt.xlim(0)
plt.ylim(0)
#plt.savefig('lienar_regression')
plt.show()
```
